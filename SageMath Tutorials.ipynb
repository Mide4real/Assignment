# Compute the limit of sin(x)/x as x tends to 0.

```python
var('x')
f(x) = sin(x)/x
lim_val = limit(f(x), x=0)
print(lim_val)
```


# Differentiate f(x) = ln(x^2 + 1)

```python
var('x')
f = log(x^2 + 1)
df = diff(f, x)
print(df)
```


# Integrate f(x) = x * e^x.

```python
var('x')
f = x * exp(x)
integral_f = integrate(f, x)
print(integral_f)
```

# Find the Taylor expansion of cos(x) around 0, up to degree 6

```python
var('x')
taylor_cos = taylor(cos(x), x, 0, 6)
print(taylor_cos)
```

Section 4.  Solving Equations and Systems

# Solve x^2 - 2 = 0 exactly

```python
var('x')
solutions = solve(x^2 - 2 == 0, x)
print(solutions)
```

```python
show(solutions)
```

# Solve exp(x) = 2 numerically.

```python
exact_sol = solve(exp(x) == 2, x)
approx_val = N(exact_sol[0].rhs())
print(approx_val)
```


# Solve the system: x + y = 5, x - y = 1

```python
var('x y')
solutions = solve([x + y == 5, x - y == 1], x, y)
print(solutions)
```

```python
show(solutions)
```

# Create a 3D plot of z = sin(x^2 + y^2) over [-2, 2] in both directions.

```python
var('x y')
f(x, y) = sin(x^2 + y^2)
plot3d(f, (x, -2, 2), (y, -2, 2))
```

# Animate the plot of sin(kx) for k = 1 to 5.

```python
var('x')
frames = [plot(sin(k*x), (x, 0, 2*pi), ymin=-1.5, ymax=1.5,
               title='sin({} * x)'.format(k)) for k in range(1, 6)]
a = animate(frames)
a.show()
```
